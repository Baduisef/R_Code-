# Introduction au package dplyr et ses verbes

R est un langage de programmation et un environnement logiciel open source spécialement conçu pour les statistiques, l'analyse  et la visualisation de données et la modélisation statistique.

Il est particulièrement populaire dans le monde de la data en raison de ses outils intégrés pour l'analyse statistique. Il offre une large gamme de fonctions statistiques et d'algorithmes pour effectuer des analyses complexes.

R est extensible grâce à un système de paquets (packages) qui permet aux utilisateurs d'ajouter des fonctionnalités spécifiques en important des packages développés par la communauté R. Dans ce cours nous allons explorer le package dplyr

## Le package dplyr:
dplyr est une grammaire de la manipulation des données, fournissant un ensemble cohérent de verbes qui aident à résoudre les défis les plus courants de la manipulation des données:

    .mutate()ajoute de nouvelles variables qui sont des fonctions des variables existantes;
    .select()choisit des variables en fonction de leurs noms.
    .filter()choisit des cas en fonction de leurs valeurs.
    .summarise()réduit les valeurs multiples jusqu'à un seul résumé.
    .arrange()change la commande des lignes.

Tous se combinent naturellement avec group_by()qui permet d'effectuer n'importe quelle opération « par groupe ».



```r
# importation des package et données
library(dplyr)
df = data(starwars)
```

### Le verbe filter:
Le verbe filter() du package dplyr est utilisé pour filtrer les lignes d'un table de données en fonction de certaines conditions spécifiées. Il permet de sélectionner les observations qui répondent à des critères spécifiques.

Voici une brève description de la fonction filter() :

    Syntaxe de base : filter(data, condition1, condition2, ...)

    Arguments :
        data : Le dataframe que vous souhaitez filtrer.
        condition1, condition2, ... : Les conditions logiques qui déterminent quelles lignes doivent être incluses dans le résultat final.

    Utilisation :
        On peut spécifier une ou plusieurs conditions à l'aide d'opérateurs de comparaison (par exemple, ==, >, <, >=, <=, !=) et d'opérateurs logiques (& pour ET, | pour OU, ! pour NON).

```r
starwars %>%
  filter(species == "Droid")
```

```r
starwars %>%
  filter(mass > 100)
```

```r
starwars %>%
  filter(species == "Droid" & mass > 100)
```

```r
starwars %>%
  filter(species == "Droid" | mass > 100)
```

### Le verbe select:
Le verbe select() du package dplyr est utilisé pour sélectionner des colonnes spécifiques d'un table de données. Il permet de créer un sous-ensemble du table de données en choisissant uniquement les variables d'intérêt.

Voici une brève description du verbe select() :

    Syntaxe de base : select(data, column1, column2, ...)

    Arguments :
        data : Le dataframe que vous souhaitez sélectionner.
        column1, column2, ... : Les noms des colonnes que vous souhaitez inclure dans le résultat final.

    Utilisation :
        Nous pouvons spécifier les colonnes à sélectionner par leur nom.
        Nous pouvons utiliser les fonctions auxiliaires de sélection, telles que starts_with(), ends_with(), contains(), matches(), pour sélectionner des colonnes qui correspondent à des motifs spécifiques.

```r
starwars %>%
  select(name, ends_with("color"))
```

```r
starwars %>%
  select(name, ends_with("color")) %>%
    filter(eye_color == "blue")
```

### Le verbe mutate:
Le verbe mutate() du package dplyr  est utilisé pour créer de nouvelles variables (colonnes) dans un table de données ou pour modifier les valeurs d'une variable existante. Il permet d'ajouter des colonnes calculées en fonction des valeurs d'autres colonnes ou d'appliquer une transformation à une colonne existante.

Voici une brève description du verbe mutate() :

    Syntaxe de base : mutate(data, new_column1 = expression1, new_column2 = expression2, ...)

    Arguments :
        data : Le dataframe sur lequel vous souhaitez effectuer la mutation.
        new_column1, new_column2, ... : Les noms des nouvelles colonnes que vous souhaitez créer.
        expression1, expression2, ... : Les expressions qui définissent les valeurs des nouvelles colonnes.

    Utilisation :
        Nous pouvez utiliser les valeurs des colonnes existantes dans les expressions pour calculer les nouvelles valeurs.
        Nous pouvez également modifier les valeurs d'une colonne existante en utilisant mutate().

```r
starwars %>%
  mutate(bmi = mass / ((height / 100)  ^ 2)) %>%
  select(name:mass, bmi)
```

```r
starwars %>%
  mutate(name, bmi = mass / ((height / 100)  ^ 2))%>%
  select(name:mass,bmi)
```

```r
starwars %>%
  mutate(name, bmi = mass / ((height / 100)  ^ 2))%>%
  select(name:mass,bmi) %>%
  arrange(desc(mass))
```

```r
starwars %>%
  mutate(grand = if_else(height > 175, "Oui", "Non")) %>%
  select(name:mass, grand)
```

### Les verbes group_by et summarise:
group_by() est utilisée pour créer des groupes dans les données, tandis que summarise() est utilisée pour calculer des statistiques sommaires à l'intérieur de ces groupes. Ces deux fonctions sont souvent utilisées en conjonction pour effectuer des opérations de résumé et d'agrégation dans des analyses de données.

```r
starwars %>%
  group_by(species) %>%
  summarise(
    n = n(),
    mass = mean(mass, na.rm = TRUE)
  )
```

```r
starwars %>%
  group_by(species) %>%
  summarise(
    n = n(),
    mass = mean(mass, na.rm = TRUE)
  ) %>%
  filter(
    n > 1,
    mass > 50
  )
```
